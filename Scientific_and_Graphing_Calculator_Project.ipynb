```python
#These import everything we need
import math
import numpy as np
from IPython.display import clear_output
import matplotlib.pyplot as plt

#The pint of this loop is to make sure that an actually useable input is inputted by the user, and if it is found to not be the correct input we want the loop will repeat itself. While I could have included the entirety of the code inside of this loop it would have been meaning less as any wrong input after this would cause an error to happen so checking the code would be pointless as if the code does not give an error that means that the input was correct.
#Only the words in parentheses work. Which is the reason why they are in parentheses in the first place.
while True:
  def asking_question_one():
    question_one = input("Do you want area of a 'triangle', 'rectangle', 'square', or 'circle'. Do you want to do a math equation write 'addition', 'subtraction', 'division' and 'multiplication'. Or if you want to draw a polynomial graph with a linear line for point of intersection write 'yes' or 'no'. If you want to graph a Trigonometric graph write 'trigonometric': ")
    #This changes the input from uppercase to lowercase which can let people input circle like cIrcLe, but when it reaches the code responsible for checking it will look like this: circle beacuse of the .lower code
    question_one = (str.lower(question_one))
    return question_one
  question_one = asking_question_one()
  if question_one not in ['triangle', 'rectangle', 'square', 'circle', 'addition', 'subtraction', 'division', 'multiplication', 'yes', 'no', 'trigonometric']:
    clear_output()
  else:
    break

#This long if/elif statment is the main bulk of the calculator, the reason why it only checks for the lowercase version of the tect is due to the fact that in the code above, I change all of the input text to it's lowercase form. Also the phrases in parentheses are the inputs that actually work.

#This finds area of a square
if question_one == "square":
  x = int(input("Please input your number. You number is "))
  y = (x * x)
  print("By having a side length of " + str(x) + " your square will have an area of " + str(y) + " squared.")

#This finds area of a triangle
elif question_one == "triangle":
  x = int(input("Please input your base. Your base is "))
  y = int(input("Please input your height. You wrote this number as your height "))
  e = ((x * y) / 2)
  print("By having " + str(x) + " as your base and " + str(y) + " as your height the triangle will have an area of " + str(e))

#This finds area of a rectangle
elif question_one == "rectangle":
  x = int(input("Please input your length. Your length is "))
  y = int(input("Please input your base. You wrote this number as your base "))
  e = (x * y)
  print("By having " + str(x) + " as your length and " + str(y) + " as your base the rectangle will have an area of " + str(e))

#This finds area of a circle
elif question_one == "circle":
  x = int(input("Please input your radus. Your radus is "))
  y = 3.141592653589793238462643383279502884197
  e = ((x * x) * y)
  print("By having " + str(x) + " as your radus your circle will have an area of " + str(e))

#This does addition
elif question_one == 'addition':
  x = int(input("Please write in your first number. "))
  y = int(input("Please write your second number. "))
  e = (x + y)
  print("By adding " + str(x) + " to " + str(y) + " we get " + str(e))

#This one does all of the subtraction
elif question_one == 'subtraction':
  x = int(input("Please write in your first number. "))
  y = int(input("Please write your second number. "))
  e = (x - y)
  print("By adding " + str(x) + " to " + str(y) + " we get " + str(e))

#This code does the division
elif question_one == 'division':
  x = int(input("Please write in your first number. "))
  y = int(input("Please write your second number. "))
  e = (x / y)
  print("By adding " + str(x) + " to " + str(y) + " we get " + str(e))

#The multiplication code is here
elif question_one == 'multiplication':
  x = int(input("Please write in your first number. "))
  y = int(input("Please write your second number. "))
  e = (x * y)
  #The reason why it's str(#) is because the values are in int format meaning you can't print them with other words unless you use commas, so by adding str(#) we are able to change the int to str which can be printed with + instead of ,
  print("By adding " + str(x) + " to " + str(y) + " we get " + str(e))

#This one if for the 'yes' answer to having both a linear line and a polynomial.
elif question_one == 'yes':
  a = int(input("What is the value of a: "))
  b = int(input("What is the value of b: "))
  p = np.linspace(-10,10,400)
  e = (a*p+b)
  plt.plot(p,e, linewidth=2.5);
  x = np.linspace(-10,10,400)
  y = input("Here is an example of a polynomial, ''3*x**2 + 4*x + 6''. Now please enter your polynomial : ")
  #This makes sure that if the symbol for by the power of is used and replace it with pythons version.
  new_y = y[0:].replace("^", "**")
  print(new_y)
  ans = eval(new_y)
  plt.plot(x,ans, linewidth=2.5)
  plt.xlabel("x-axis")
  plt.ylabel("y-axis")
  plt.title("This is the graph of the polynomial " + str(new_y) + " and the intersection with the linear line of (" + str(a) + ")x + (" + str(b) + ")")
  plt.grid()
  plt.style.use('dark_background')
  plt.show()

#This one is here for if you have a polynomial, but no linear line to graph.
elif question_one == 'no':
  x = np.linspace(-10,10,400)
  y = input("Here is an example of a polynomial, ''3*x**2 + 4*x + 6''. Now please enter your polynomial: ")
  #This makes sure that if the symbol for by the power of is used and replace it with pythons version.
  new_y = y[0:].replace("^", "**")
  print(new_y)
  ans = eval(new_y)
  plt.plot(x,ans, linewidth=5)
  plt.xlabel("x-axis")
  plt.ylabel("y-axis")
  plt.title("This is the graph of the polynomial " + str(new_y))
  plt.grid()
  plt.style.use('dark_background')
  plt.show()

#And this one is for running the trigonometric code for graphing.
elif question_one == 'trigonometric':
  while True:
    # This function asks the user to choose the trigonometry function
    def trigonometry_function():
      user_input = input("Make a choice from 'sin', 'cos' or 'tan': ")
      user_input = (str.lower(user_input))
      return user_input
    # This gets the user's choice
    user_input = trigonometry_function()
    # This checks if the users input choice is correct. If it is incorrect the code above will be repeated
    if user_input not in ['sin', 'cos', 'tan']:
      clear_output()
    else:
      break
  # This generates the values for X
  x = np.linspace(-2*np.pi, 2*np.pi, 100)
  # This defines Y values based on the user's choice
  if user_input == 'sin':
    y = np.sin(x)
  elif user_input == 'cos':
    y = np.cos(x)
  elif user_input == 'tan':
    y = np.tan(x)
  # This prevents the lines from connecting to each other and seperates them
  y[:-1][abs(np.diff(y)) > 10 ] = np.nan
  # This plots the x and y values
  plt.figure(figsize=(10, 5))
  plt.plot(x, y, label = user_input, color='green')
  plt.xlabel('x values')
  plt.ylabel(user_input +'(x)')
  plt.legend()
  plt.grid()
  plt.title("Graph of " + user_input)
  plt.show()
```
